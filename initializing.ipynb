```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data for Reliance Industries
df = pd.read_csv('data/RELIANCE_minute.csv')

# Convert the 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])

# Set 'date' as the index
df.set_index('date', inplace=True)

# Sort chronologically just in case
df.sort_index(inplace=True)

# Let's look at the first few rows
df.head()
```

```python
# Feature Engineering: Generating Technical Indicators

# 1. Simple Moving Average (SMA) - 50 periods (50 minutes)
df['SMA_50'] = df['close'].rolling(window=50).mean()

# 2. Exponential Moving Average (EMA) - 20 periods
df['EMA_20'] = df['close'].ewm(span=20, adjust=False).mean()

# 3. Relative Strength Index (RSI) - 14 periods
delta = df['close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df['RSI_14'] = 100 - (100 / (1 + rs))

# Drop NA values generated by the rolling windows
df.dropna(inplace=True)

# Let's plot the close price along with SMA and EMA for a small slice of data (e.g., 300 minutes)
subset = df.iloc[500:800] # Selecting an arbitrary day slice after the first day
plt.figure(figsize=(14, 7))
plt.plot(subset.index, subset['close'], label='Close Price', color='black', alpha=0.6)
plt.plot(subset.index, subset['SMA_50'], label='SMA 50', color='blue', linestyle='--')
plt.plot(subset.index, subset['EMA_20'], label='EMA 20', color='red', linestyle='-.')
plt.title('Reliance Minute-Level Data - Price vs Moving Averages')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
